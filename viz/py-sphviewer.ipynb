```python
%pylab inline
import sphviewer
from pygadgetreader import *
```

```python
positions = readsnap("../../ICs/GalIC/MW/chervin_010", "pos", "dm")
mass = readsnap("../../ICs/GalIC/MW/chervin_010", "mass", "dm")
```

```python
plt.scatter(positions[:,0], positions[:,1], s=1.0)
```

```python
Particles = sphviewer.Particles(positions.T,mass)
```

```python
Scene = sphviewer.Scene(Particles)
```

```python
Render = sphviewer.Render(Scene)
```

```python
Scene.update_camera(r=4000.00)
#Scene.update_camera(r='infinity')
Render = sphviewer.Render(Scene)
Render.set_logscale()
img = Render.get_image()
extent = Render.get_extent()
fig = plt.figure(1,figsize=(5,5))
ax1 = fig.add_subplot(111)
ax1.imshow(img, extent=extent, origin='lower', cmap='winter', vmax=0.000001)
ax1.set_xlabel('X', size=15)
ax1.set_ylabel('Y', size=15)
```
