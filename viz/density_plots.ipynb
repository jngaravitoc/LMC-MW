```python
for i in range(100):
    ds = yt.load(('../data/LMCMW/MWmLMC3/Hernquist/H12/MWfLMC3H12_{:03d}.hdf5').format(i), bounding_box=bbox)
    ds.index
    ad= ds.all_data()
    p = yt.ParticlePlot(ds, ('PartType2','particle_position_y'), ('PartType2','particle_position_z'), ('PartType2','particle_mass'))
    p.set_xlim(-37, 37)
    p.set_ylim(-37, 37)
    p.set_unit('particle_position_y','kpc')
    p.set_unit('particle_position_z','kpc')
    p.set_unit('particle_mass', 'Msun')
    p.annotate_text(-35, 30, ('T={:02f}Gyrs').format(i/10.0))
    #p.show()
    p.save('YZMWdisk_{:03d}.png'.format(i))
```

```python
for i in range(100):
    ds = yt.load(('../data/LMCMW/MWmLMC3/Hernquist/H12/MWfLMC3H12_{:03d}.hdf5').format(i), bounding_box=bbox)
    ds.index
    ad= ds.all_data()
    p1 = yt.ParticlePlot(ds, ('PartType2','particle_position_x'), ('PartType2','particle_position_y'), ('PartType2','particle_mass'))
    p2 = yt.ParticlePlot(ds, ('PartType2','particle_position_y'), ('PartType2','particle_position_z'), ('PartType2','particle_mass'))

    p1.set_xlim(-30, 30)
    p1.set_ylim(-30, 30)
    
    p2.set_xlim(-15, 15)
    p2.set_ylim(-30, 30)
    #p.show()
    p1.save('disk_{:03d}.png'.format(i))
    p2.save('YZdisk_{:03d}.png'.format(i))
```
