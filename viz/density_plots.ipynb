```python
dt = 0.1
for i in range(150):
    ds = yt.load(('../data/LMCMW/MWmLMC6/H1/MWfLMC6H1_{:03d}.hdf5').format(i), bounding_box=bbox)
    ds.index
    ad= ds.all_data()
    p = yt.ParticlePlot(ds, ('PartType1','particle_position_y'), ('PartType1','particle_position_z'), ('PartType1','particle_mass'))
    p.set_xlim(-370, 370)
    p.set_ylim(-370, 370)
    p.set_unit('particle_position_y','kpc')
    p.set_unit('particle_position_z','kpc')
    p.set_unit('particle_mass', 'Msun')
    p.annotate_text(-350, 300, ('T={:02f}Gyrs').format(dt), color='y')
    #p.show()
    dt += 0.1
    p.save('LMC6/LMC6MW_{:03d}.png'.format(i))
```
