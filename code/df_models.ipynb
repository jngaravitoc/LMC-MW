```python
# Function that reads the data from the Analytic integration
#Galactocentic distances and velocities
def reading_data(snap_name):
    host = loadtxt('../../orbits/data/lmc/'+ snap_name +'_host.txt')
    sat = loadtxt('../../orbits/data/lmc/'+ snap_name +'_sat.txt')
    t = host[:,0] 
    # Defining galactocentric distances and velocities
    x = sat[:,1] - host[:,1] 
    y = sat[:,2] - host[:,2] 
    z = sat[:,3] - host[:,3] 

    vx = sat[:,4] - host[:,4] 
    vy = sat[:,5] - host[:,5] 
    vz = sat[:,6] - host[:,6]        

    Rgal = np.sqrt(x**2.0 + y**2.0 + z**2.0)
    Vgal = np.sqrt(vx**2.0 + vy**2.0 + vz**2.0)
    
    return t, x, y, z, vx, vy, vz, Rgal, Vgal
```

```python
tgal1, xgal1, ygal1, zgal1, vxgal1, vygal1, vzgal1, Rgal1, Vgal1 = reading_data('lmc1_Herndf_vdm_cl1')
tgal2, xgal2, ygal2, zgal2, vxgal2, vygal2, vzgal2, Rgal2, Vgal2 = reading_data('lmc3_Herndf_vdm_cl1')
tgal3, xgal3, ygal3, zgal3, vxgal3, vygal3, vzgal3, Rgal3, Vgal3 = reading_data('lmc6_Herndf_vdm_cl1')

tgal4, xgal4, ygal4, zgal4, vxgal4, vygal4, vzgal4, Rgal4, Vgal4 = reading_data('lmc1Herndf_vdm_cl2')
tgal5, xgal5, ygal5, zgal5, vxgal5, vygal5, vzgal5, Rgal5, Vgal5 = reading_data('lmc3Herndf_vdm_cl2')
tgal6, xgal6, ygal6, zgal6, vxgal6, vygal6, vzgal6, Rgal6, Vgal6 = reading_data('lmc6_Herndf_vdm_cl2')

tgal7, xgal7, ygal7, zgal7, vxgal7, vygal7, vzgal7, Rgal7, Vgal7 = reading_data('lmc1_Herndf_vdm_cl3')
tgal8, xgal8, ygal8, zgal8, vxgal8, vygal8, vzgal8, Rgal8, Vgal8 = reading_data('lmc3_Herndf_vdm_cl3')
tgal9, xgal9, ygal9, zgal9, vxgal9, vygal9, vzgal9, Rgal9, Vgal9 = reading_data('lmc6_Herndf_vdm_cl3')

#tgal1, xgal1, ygal1, zgal1, vxgal1, vygal1, vzgal1, Rgal1, Vgal1 = reading_data('lmc6_Herndf_vdm_cl2')
#tgal1, xgal1, ygal1, zgal1, vxgal1, vygal1, vzgal1, Rgal1, Vgal1 = reading_data('lmc6_Herndf_vdm_cl2')
#tgal1, xgal1, ygal1, zgal1, vxgal1, vygal1, vzgal1, Rgal1, Vgal1 = reading_data('lmc6_Herndf_vdm_cl2')
```
